```python
n=int(input())

for i in range (1, n+1):
  for j in range (1,n+1):
    if(i==1 or i==n or j==1):
        print("C",end =" ")
    else:
        print(" ",end =" ") # Print space for inner parts
  print()
```

```python
n=5
for i in range (1, n + 1):
    ch=65
    for k in range (n-i):
      print(" ", end="")
    for s in range (i):
      print(chr(ch),end =" ")
      ch=ch+1
    print()
```

```python
n=5
for i in range (n, 0, -1):
    ch=65
    for k in range (n-i):
      print(" ", end="")
    for s in range (i):
      print(chr(ch),end =" ")
      ch=ch+1
    print()
```

```python
n=int(input())

for i in range (1, n+1):
  for j in range (1,n+1):
    if( j==i):
        print(i, end =" ")
    else:
        print(" ",end =" ")
  print()

  for j in range (n, 0, -1):
    if( j==i):
        print(i, end =" ")
    else:
        print(" ",end =" ")
  print()
```
